```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#I chose to examine the UN data regarding:
#Youth literacy rate, population 15-24 years, gender parity index.
df = pd.read_csv('Unit_2_Challenge_Data/UNYouthLitRate15-24.csv')

#Creating a dataframe for showing the literacy rate of the least developed countries (1985-2015)
least_developed = df.iloc[678:709,5]

#Plotting the values in chronological order (1985-2015) instead of (2015-1985) as it appears in the csv file.
plt.plot(least_developed.values[::-1])

#Setting the plot titles and labels.
plt.title('Literacy Rates of 15-24 year olds in MDG Least Developed Countries: 1985-2015')
plt.ylabel('Literacy Rate')
plt.xlabel('Year')

plt.show()
```

```python
#Next, I will overlay data about Sub-Saharan African (SSA) Countries over the plot above.
#Creating a dataframe for SSA countries.
ssa = df.iloc[875:906,5]

#Plotting the values in chronological order (1985-2015) instead of (2015-1985) as it appears in the csv file.
plt.plot(ssa.values[::-1], color="red")

#Add MDG Least Developed Countries
plt.plot(least_developed.values[::-1], color="blue")

#Setting the plot titles and labels.
plt.title('Literacy Rates of 15-24 year olds in MDG Least Developed Countries (Blue) vs. SSA Countries (Red): 1985-2015')
plt.ylabel('Literacy Rate')
plt.xlabel('Year')

plt.show()
```

```python
#ANALYSIS: The plot above shows that literacy rates in SSA countries were higher than in the least developed countries
#in year 0 (1985) and remained higher until approximately year 20 (2005).  At this point, the literacy rates in SSA 
#countries were still increasing, but surpassed by the least developed countries. This trend continued until 2015, 
#the last year for which this data was available.
```

```python
#Next, I will take a look at two example counties belonging to each category in the above plot.  Cameroon is not
#listed among the MDG Least Developed Countries (according to: http://unohrlls.org/custom-content/uploads/2015/08/The-Least-Developed-Countries-Report-2014.pdf,
#page 3), but it is an SSA country according to (https://www.loc.gov/rr/amed/guide/afr-countrylist.html). Similarly, 
#Bangladesh is listed as an MDG Least Developed Country, but it is not located in SSA. In particular, I am looking
#at education expenses after 2005, when the trend in literacy rates shifted between these two regions.

df2 = pd.read_csv('Unit_2_Challenge_Data/UNGovtExpEdPercentTotalExp.csv')

#Creating a dataframe for showing Cameroon's education expenditures as a percent of total government expenditures from 
#2006-2009 (the years for which this data is available for both countries).

cameroon_ed_exp = df2.iloc[286:290,5]

#Plotting the values in chronological order (2006-2009) instead of (2009-2006) as it appears in the csv file.
plt.plot(cameroon_ed_exp.values[::-1])

#Setting the plot titles and labels.
plt.title("Education Expenses as a Percent of Total Gov't Expenses (Cameroon): 2006-2009")
plt.ylabel('Percent')
plt.xlabel('Year')

plt.show()
```

```python
#Next, I will overlay data from Bangladesh over the plot above.

#Creating a dataframe for Bangladesh.
bangladesh_ed_exp = df2.iloc[114:118,5]

#Plotting the values in chronological order (2006-2009) instead of (2006-2009) as it appears in the csv file.
plt.plot(bangladesh_ed_exp.values[::-1], color="blue")

#Add Cameroon to the plot.
plt.plot(cameroon_ed_exp.values[::-1], color="red")

#Setting the plot titles and labels.
plt.title("Education Expenses as a Percent of Total Gov't Expenses (Cameroon[Red] vs. Bangladesh[Blue]): 2006-2009")
plt.ylabel('Percent')
plt.xlabel('Year')

plt.show()
```

```python
#In order to more easily visualize the possible relationship between gov't expenditures on education and literacy
#outcomes, I will create a subplot showing both plots for data between 2006 and 2009.

#First, I need to recreate the comparison of youth literacy rates to focus in on 2006 - 2009.

#Re-creating a dataframe for SSA countries.
ssa = df.iloc[881:885,5]

#Re-creating a dataframe for Least Developed countries.
least_developed = df.iloc[684:688,5]

#Plotting the values in chronological order (2006-2009) instead of (2009-2006) as it appears in the csv file.
plt.plot(ssa.values[::-1], color="red")

#Add MDG Least Developed Countries
plt.plot(least_developed.values[::-1], color="blue")

#Setting the plot titles and labels.
plt.title('Literacy Rates of 15-24 year olds in MDG Least Developed Countries (Blue) vs. SSA Countries (Red): 2006-2009')
plt.ylabel('Literacy Rate')
plt.xlabel('Year')

plt.show()
```

```python
#Here I am rendering the subplot.

plt.figure(figsize=(18,6))

#First, the literacy rate comparison -------------------------------------------------------------------------------
plt.subplot(1,2,1)

#Plotting the values in chronological order (2006-2009) instead of (2009-2006) as it appears in the csv file.
plt.plot(ssa.values[::-1], color="red")

#Add MDG Least Developed Countries
plt.plot(least_developed.values[::-1], color="blue")

#Setting the plot titles and labels.
plt.title('Youth Literacy Rates: Least Developed Countries (Blue) vs. SSA Countries (Red): 2006-2009')
plt.ylabel('Literacy Rate')
plt.xlabel('Year')

#Second, the education expenditure comparison ----------------------------------------------------------------------
plt.subplot(1,2,2)

#Plotting the values in chronological order (2006-2009) instead of (2006-2009) as it appears in the csv file.
plt.plot(bangladesh_ed_exp.values[::-1], color="blue")

#Add Cameroon to the plot.
plt.plot(cameroon_ed_exp.values[::-1], color="red")

#Setting the plot titles and labels.
plt.title("Education Expenses (Cameroon[Red] vs. Bangladesh[Blue]): 2006-2009")
plt.ylabel('Percent')
plt.xlabel('Year')

plt.show()
```

```python
#The comparisons above seems to show that the greater variance in government expenditures may be related to 
#poor outcomes in terms of literacy rates.  In particular, over the period 2006 - 2008, Cameroon's expenditures on education 
#fluctuated a high of 21.5% of total expenses to a low of 15.5% of total expenses (a variance of 6 percentage points).
#While Bangladesh's expenses on education were generally lower than Cameroon's, its variance during this period was
#within one percentage point. During the period Bangladesh's youth literacy rates were higher than Cameroon's. 
```
